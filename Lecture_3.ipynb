```python
text = ''' 
This is undoubtedly a classic game that will keep your team excited for weeks! Draw names from a hat and have each person buy a small gift for their Secret Santa. Set a spending limit (i.e. £20) and have everyone bring their wrapped gift to the party. Then, take turns opening them up in front of everyone. This is a great way to spread some holiday cheer without breaking the bank.

By introducing themes like “Sustainable Gifts” or “Gifts that Spark Joy”, the experience can be elevated. Colleagues can engage in guessing games, trying to determine who their Santa was based on the nature of the gift. The anticipation, the unveiling, and the stories that come with it not only add joy but also deepen bonds among team members, making it an indispensable Christmas idea.
'''
```

```python
text = text.lower()
for char in ['.',',','“','”','\n','!']:
    text = text.replace(char,' ')
print(text)
```

```python
words = text.split()
words 
```

```python
fre_words = [] #[[fre,word],....]
for w in set(words):
    fre_words.append([words.count(w),w])
fre_words
```

```python
sorted(fre_words,reverse= True)
```
